```
# Exercicio 3:
numero1 = float(input('Digite o primeiro número: '))
numero2 = float(input('Digite o segundo número: '))
soma = numero1 + numero2
print(f'A soma é {soma}')
```

```
# Exercicio 4:
nota1 = float(input('Digite a primeira nota: '))
nota2 = float(input('Digite a segunda nota: '))
nota3 = float(input('Digite a terceira nota: '))
nota4 = float(input('Digite a quarta nota: '))
media = (nota1 + nota2 + nota3 + nota4) / 4
print(f'A média é {media}')
```

```
# Exercicio 5:
metros = float(input('Digite a quantidade de metros: '))
centimetros = metros * 100
print(f'{metros} metros são {centimetros} centimetros')
```

```
# Exercicio 6:
import math
raio = float(input('Digite o raio do círculo: '))
area = math.pi * raio**2
print(f'A área do círculo é {area}')
```

```
# Exercicio 7:
lado = float(input('Digite o lado do quadrado: '))
area = lado ** 2
dobro_area = area * 2
print(f'A área do quadrado é {area} e o dobro dessa área é {dobro_area}')
```

```
# Exercicio 8:
ganho_hora = float(input('Quanto você ganha por hora? '))
horas_trabalhadas = float(input('Número de horas trabalhadas no mês: '))
salario = ganho_hora * horas_trabalhadas
print(f'Seu salário neste mês será de {salario}')
```

```
# Exercicio 9:
fahrenheit = float(input('Digite a temperatura em graus Fahrenheit: '))
celsius = 5 * ((fahrenheit - 32) / 9)
print(f'A temperatura em Celsius é {celsius}')
```

```
# Exercicio 10:
celsius = float(input('Digite a temperatura em graus Celsius: '))
fahrenheit = (celsius * 9/5) + 32
print(f'A temperatura em Fahrenheit é {fahrenheit}')
```

```
# Exercicio 11:
numero1 = int(input('Digite o primeiro número inteiro: '))
numero2 = int(input('Digite o segundo número inteiro: '))
numero3 = float(input('Digite um número real: '))
resultado1 = (2 * numero1) * (numero2 / 2)
resultado2 = (3 * numero1) + numero3
resultado3 = numero3 ** 3
print(f'O produto do dobro do primeiro com metade do segundo é {resultado1}')
print(f'A soma do triplo do primeiro com o terceiro é {resultado2}')
print(f'O terceiro elevado ao cubo é {resultado3}')
```

```
# Exercicio 12:
altura = float(input('Digite sua altura: '))
peso_ideal = (72.7 * altura) - 58
print(f'Seu peso ideal é {peso_ideal} kg')
```

```
# Exercicio 13:
altura = float(input('Digite sua altura: '))
peso_ideal_homem = (72.7 * altura) - 58
peso_ideal_mulher = (62.1 * altura) - 44.7
print(f'Seu peso ideal é {peso_ideal_homem} kg para homens')
print(f'Seu peso ideal é {peso_ideal_mulher} kg para mulheres')
```
